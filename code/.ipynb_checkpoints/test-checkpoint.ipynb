Import libraries and dependencies.

```python
%matplotlib inline
from __future__ import print_function
import matplotlib.pyplot as plt
import tensorflow as tf
import scipy.io as io
import numpy as np
import csv

import os
from datetime import datetime
```

Load data.

```python
train_data = io.loadmat("../data/train_32x32.mat")
test_data = io.loadmat("../data/test_32x32.mat")

train_dataset = train_data['X']
train_labels = train_data['y']
test_dataset = test_data['X']
test_labels = test_data['y']

print('Training set', train_dataset.shape, train_labels.shape)
print('Test set', test_dataset.shape, test_labels.shape)    
```

Reformat data to a friendly shape to Tensorflow.

```python
IMAGE_SIZE = 32
NUM_CLASSES = 10
NUM_CHANNELS = 1
NUM_TRAIN_DATA = 73000

def reformat(dataset, labels):
    dataset = np.transpose(dataset, (3, 0, 1, 2)).astype(np.float32)
    one_hot_labels = []
    for label in labels:
        one_hot_label = np.zeros(NUM_CLASSES, np.float32)
        if label[0] == 10:
            one_hot_label[0] = 1.0
        else:
            one_hot_label[label[0]] = 1.0
        one_hot_labels.append(one_hot_label)
    labels = np.array(one_hot_labels).astype(np.float32)
    return dataset, labels

valid_dataset = train_dataset[:, :, :, NUM_TRAIN_DATA : ]
valid_labels = train_labels[NUM_TRAIN_DATA : ]
train_dataset = train_dataset[:, :, :, : NUM_TRAIN_DATA]
train_labels = train_labels[ : NUM_TRAIN_DATA]

train_dataset, train_labels = reformat(train_dataset, train_labels)
valid_dataset, valid_labels = reformat(valid_dataset, valid_labels)
test_dataset, test_labels = reformat(test_dataset, test_labels)

print('Training set', train_dataset.shape, train_labels.shape)
print('Validation set', valid_dataset.shape, valid_labels.shape)
print('Test set', test_dataset.shape, test_labels.shape)
```

Show input images.

```python
plt.rcParams['figure.figsize'] = (15.0, 15.0)
_, ax = plt.subplots(nrows=1, ncols=10)

for i, j in enumerate(np.random.randint(0, train_labels.shape[0], size=10)):
    ax[i].axis('off')
    ax[i].set_title(np.argmax(train_labels[j]), loc='center')
    ax[i].imshow(train_dataset[j])
```

Normalization operation.

```python
def normalize(dataset):
    new_dataset = np.add.reduce(dataset, keepdims=True, axis=3)
    new_dataset = new_dataset / 3.0
    return new_dataset / 128.0 - 1.0

new_train_dataset = normalize(train_dataset)
new_valid_dataset = normalize(valid_dataset)
new_test_dataset = normalize(test_dataset)

print('Training set', new_train_dataset.shape, train_labels.shape)
print('Validation set', new_valid_dataset.shape, valid_labels.shape)
print('Test set', new_test_dataset.shape, test_labels.shape)
```

Plot normalized images.

```python
_, ax = plt.subplots(nrows=1, ncols=10)
for i, j in enumerate(np.random.randint(0, train_labels.shape[0], size=10)):
    ax[i].axis('off')
    ax[i].set_title(np.argmax(train_labels[j]), loc='center')
    ax[i].imshow(new_train_dataset[j].squeeze())
```

```python
def accuracy(predictions, labels):
    s = 0
    for i in range(len(predictions)):
        if predictions[i] == np.argmax(labels[i]):
            s += 1
    return 100 * s / len(predictions)
```

Set hyper-parameters.

```python
batch_size = 64
conv1_depth = 16
conv2_depth = 32
hidden_layer_size = 64
patch_size = 5
start_learning_rate = 0.001
regularization_lambda = 0.0005
dropout_rate = 0.9
learning_rate_decay_rate = 0.99
```

ConvNet setting.

```python
num_sample = 100

graph = tf.Graph()
with graph.as_default():
    tf_test_dataset = tf.placeholder(tf.float32, shape=(num_sample, IMAGE_SIZE, IMAGE_SIZE, NUM_CHANNELS))

    conv1_filter = tf.Variable(tf.truncated_normal([patch_size, patch_size, NUM_CHANNELS, conv1_depth], stddev = 0.1))
    conv1_biases = tf.Variable(tf.zeros([conv1_depth]))

    conv2_filter = tf.Variable(tf.truncated_normal([patch_size, patch_size, conv1_depth, conv2_depth], stddev = 0.1))
    conv2_biases = tf.Variable(tf.constant(1.0, shape = [conv2_depth]))
    
    layer3_weights = tf.Variable(tf.truncated_normal([IMAGE_SIZE // 4 * IMAGE_SIZE // 4 * conv2_depth, hidden_layer_size], stddev = 0.1))
    layer3_biases = tf.Variable(tf.constant(1.0, shape = [hidden_layer_size]))
    
    layer4_weights = tf.Variable(tf.truncated_normal([hidden_layer_size, NUM_CLASSES], stddev = 0.1))
    layer4_biases = tf.Variable(tf.constant(1.0, shape = [NUM_CLASSES]))

    def model(dataset, train = False):
        conv = tf.nn.conv2d(dataset, conv1_filter, [1, 1, 1, 1], padding = 'SAME')
        hidden = tf.nn.relu(conv + conv1_biases)
        hidden = tf.nn.max_pool(hidden, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')

        conv = tf.nn.conv2d(hidden, conv2_filter, [1, 1, 1, 1], padding = 'SAME')
        hidden = tf.nn.relu(conv + conv2_biases)
        hidden = tf.nn.max_pool(hidden, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
    
        shape = hidden.get_shape().as_list()
        reshape = tf.reshape(hidden, [shape[0], shape[1] * shape[2] * shape[3]])
        hidden = tf.nn.relu(tf.matmul(reshape, layer3_weights) + layer3_biases)
        
        if train:
            hidden = tf.nn.dropout(hidden, dropout_rate)

        return tf.matmul(hidden, layer4_weights) + layer4_biases

    logits = model(tf_test_dataset, False)
    
    prediction = tf.nn.softmax(model(tf_test_dataset))
    test_prediction = tf.argmax(prediction, 1)
    
    saver = tf.train.Saver()
```

```python
def get_samples(dataset, labels):
    indices = []
    sample_dataset = []
    sample_labels = []
    
    for i in range(10):
        picked = np.zeros(10, int)
        count = np.zeros(10, int)
        cnt = 0
        while cnt != 10:
            idx = np.random.randint(0, dataset.shape[0] - 1)
            lab = np.argmax(labels[idx])
            if count[lab] == 0:
                count[lab] = 1
                picked[lab] = idx
                cnt += 1
        
        for d in picked:
            sample_dataset.append(dataset[d])
            sample_labels.append(labels[d])
            indices.append(d)
    return np.array(sample_dataset), np.array(sample_labels), np.array(indices)
```

```python
model_path = "model/1_20.ckpt"

with tf.Session(graph = graph) as session:
    tf.initialize_all_variables().run()
    saver.restore(session, model_path)
    sample_dataset, sample_labels, idx = get_samples(new_test_dataset, test_labels)
    feed_dict = {tf_test_dataset : sample_dataset}
    test_pred = session.run(test_prediction, feed_dict = feed_dict)
    print(test_pred)
    print(accuracy(test_pred, sample_labels))     
```
